# Gemini Vision API com Output Estruturado

Neste notebook, vamos usar a **nova SDK `google-genai`** para enviar uma imagem e receber uma resposta estruturada em JSON (utilizando Pydantic para definir o esquema).

```python
import os
import json
from google import genai
from pydantic import BaseModel
from dotenv import load_dotenv
from PIL import Image
import IPython.display

# Carregar variáveis de ambiente
load_dotenv()

api_key = os.getenv("GEMINI_API_KEY") or os.getenv("GOOGLE_API_KEY")

if not api_key:
    print("ERRO: Variável de API não encontrada.")
else:
    # Inicializar o cliente com a nova SDK
    client = genai.Client(api_key=api_key)
    print("Cliente Google GenAI inicializado com sucesso!")
```

## 1. Definir o Esquema de Resposta

Vamos usar uma classe Pydantic para definir exatamente como queremos que o Gemini responda.

```python
class ImageAnalysis(BaseModel):
    objeto_principal: str
    descricao_detalhada: str
    cores_predominantes: list[str]
    sentimento_da_imagem: str
    elementos_secundarios: list[str]
    localizacao_provavel: str
```

## 2. Selecionar e Visualizar Imagem

```python
img_path = "imagens/bird.jpg"
img = Image.open(img_path)
IPython.display.display(img)
```

## 3. Gerar Output Estruturado

Agora vamos pedir ao Gemini para analisar a imagem e retornar os dados seguindo o esquema que definimos.

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=["Analise esta imagem e retorne os dados no formato estruturado solicitado.", img],
    config={
        "response_mime_type": "application/json",
        "response_schema": ImageAnalysis,
    }
)

# Converter o JSON de resposta para o objeto Pydantic
analysis_result = response.parsed

print("---" * 10)
print("ANÁLISE ESTRUTURADA:")
print(json.dumps(analysis_result.model_dump(), indent=2, ensure_ascii=False))
```

### Acessando os dados de forma programática

Como usamos Output Estruturado, podemos acessar os campos facilmente:

```python
print(f"Objeto: {analysis_result.objeto_principal}")
print(f"Cores: {', '.join(analysis_result.cores_predominantes)}")
```
